```python
for model, scenarios in scenario_mapping.items():
    scenario_list = list()
    for s in scenarios:
        df = conn.query(model=model, scenario=s)

        if df.empty:
            print(f"No data for {model}|{s}")
            continue
        
        df.set_meta("https://doi.org/10.5281/zenodo.10032658", "DOI")
        df.set_meta("Pietzcker et al., 2023, ZENODO", "Citation")
        df.set_meta("WP1", "Work package")
        df.set_meta("Diagnostic", "Scenario type")

        if "World" in df.region:
            df.set_meta("yes", "coverage: global")
        if "OECD & EU (R5)" in df.region:
            df.set_meta("yes", "coverage: R5")
        if "Europe (R10)" in df.region:
            df.set_meta("yes", "coverage: R10")
        if "EU27 & UK" in df.region:
            df.set_meta("yes", "coverage: EU27 & UK")
        if "EU27" in df.region:
            df.set_meta("yes", "coverage: EU27")
        if set(df.region).issubset(countries):
            df.set_meta("yes", "coverage: country")

        df.rename(model={"IMAGE 3.2": "IMAGE 3.3", "MESSAGEix-GLOBIOM 1.2": "MESSAGEix-GLOBIOM 1.1"}, inplace=True)

        if model == "PRIMES 2022":
            df.filter(region=["EU27", "EU27 & UK"], inplace=True)
        
        scenario_list.append(df)

    pyam.concat(scenario_list).to_excel(f"ECEMF_diagnostics_v2.0_{model.replace(" ", "_")}.xlsx")
    
    print(f"Done exporting {model} results")
```
